```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
from math import ceil, floor
```

```python
def thresholding(f, L):
    # create a new image with zeros
    f_tr = np.ones(f.shape).astype(np.uint8)
    # setting to 1 the pixels above the threshold
    f_tr[np.where(f < L)] = 0
    return f_tr

def otsu_threshold(img, max_L):
    
    M = np.product(img.shape)
    min_var = []
    hist_t,_ = np.histogram(img, bins=256, range=(0,256))
    
    img_t = thresholding(img, 0)
    
    for L in np.arange(1, max_L):
        img_ti = thresholding(img, L)
        # computing weights
        w_a = np.sum(hist_t[:L])/float(M)
        w_b = np.sum(hist_t[L:])/float(M)
        # computing variances
        sig_a = np.var(img[np.where(img_ti == 0)])
        sig_b = np.var(img[np.where(img_ti == 1)])
        
        min_var = min_var + [w_a*sig_a + w_b*sig_b]
        
    img_t = thresholding(img, np.argmin(min_var))
    
    return img_t, np.argmin(min_var)

def normalization(g):
    g = (((g - g.min())* 255)/(g.max() - g.min()))
    return g
```

```python
image1 = imageio.imread('Images/Orion/photo1.jpg', as_gray=True)
image2 = imageio.imread('Images/photo2.jpg', as_gray=True)
image3 = imageio.imread('Images/Cancer/con_CNC_001.png', as_gray=True)



hist1,_ = np.histogram(image1, bins=256, range=(0,256))
#hist1 = histogram(image1, 265)
#plt.subplot(221); plt.bar(range(256), hist1)

hist2,_ = np.histogram(image2, bins=256, range=(0,256))
#hist2 = histogram(image2, 256)
#plt.subplot(222); plt.bar(range(256), hist2)

print(np.mean(image2))

T = ceil(8/floor(np.log(np.mean(image1))))*np.mean(image1)
print(T)
img_o1 = thresholding(image1,T)

W = 3.9
T = W*np.mean(image2)
print(T)
img_o2 = thresholding(image2,T)

W = 4.3
T = W*np.mean(image3)
print(T)
img_o3 = thresholding(image3,T)



plt.figure(figsize=(10,10))
plt.subplot(321); plt.imshow(image1, cmap="gray"); plt.axis('off'); plt.title('Original Image')
plt.subplot(322); plt.imshow(img_o1, cmap="gray"); plt.axis('off'); plt.title('thresholded image')
plt.subplot(323); plt.imshow(image2, cmap="gray"); plt.axis('off'); plt.title('Original Image')
plt.subplot(324); plt.imshow(img_o2, cmap="gray"); plt.axis('off'); plt.title('thresholded image')
plt.subplot(325); plt.imshow(image3, cmap="gray"); plt.axis('off'); plt.title('Original Image')
plt.subplot(326); plt.imshow(img_o3, cmap="gray"); plt.axis('off'); plt.title('thresholded image')
```
